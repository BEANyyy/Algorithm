TSV 파일 열어서 딕셔너리형으로 변환하기

```python
import csv

# TSV 파일 열기
with open('address_book2020.tsv', 'r', encoding='utf-8', newline='') as tsv_file:
    tsv_reader = csv.reader(tsv_file, delimiter='\t')

    # 첫 번째 줄 : 헤더
    headers = next(tsv_reader)

    # 딕셔너리 초기화
    data = {}

    # 각 행을 딕셔너리로 변환
    for row in tsv_reader:
        keys = row[0]
        values = row[1:]
        data[keys] = dict(zip(headers[1:], values))

# 결과 출력
#print(data)

#dict_keys를 리스트로 형변환
keys = [*data.keys()]
print(keys)
```

Insert

```python
def create_node(key):
    return {'key': key, 'left': None, 'right': None, 'p': None}

def insert(T, z):   # z : 삽입하고자 하는 노드
    # 'key' : 노드의 키 값, 'p' : 부모 노드의 값
    y = None
    # T 딕셔너리에서 'root'라는 키를 가져와서 x에 할당
    x = T.get('root')   # x : 현재 노드

    # x 값이 null이 아닐 때
    while x is not None:
        y = x
        if z.get('key') < x.get('key'): # 삽입하고자 하는 노드(z)가 root(x) 보다 작을 때
            x = x.get('left')           # 왼쪽에 삽입
        else:
            x = x.get('right')          # 반대 : 오른쪽에 삽입
    z['p'] = y                          # 부모 노드를 y로 설정
    if y is None:                       # y가 비어있다면
        T['root'] = z                   # z를 root로 설정
    elif z.get('key') < y.get('key'):   # y가 비어 있지 않은 경우 z와 비교하여 삽입
        y['left'] = z
    else:
        y['right'] = z

tree = {'root':None}
for i in keys:
    node = create_node(i)
    insert(tree, node)

print(tree)
```

```python
def delete(root, key):
    # 노드 검색
    node = search_node(root, key)
    
    if node is None:
        return None
    
    # 삭제할 노드와 자식 노드 개수 판별
    if node['left'] is None or node['right'] is None:
        y = node
    else:
        y = get_successor(node)
    
    # 자식 노드 설정
    if y['left'] is not None:
        x = y['left']
    else:
        x = y['right']
    
    if x is not None:
        x['p'] = y['p']
        
    # y가 루트 노드인 경우
    if y['p'] is None:
        root = x
    else:
        # y의 부모 노드와 자식 노드 연결
        if y == y['p']['left']:
            y['p']['left'] = x
        else:
            y['p']['right'] = x
        
    # y가 삭제될 노드가 아닌 경우
    if y != node:
        node['key'] = y['key']
        node['satellite_data'] = y['satellite_data']
        
    return y

def search_node(node, key):
    if node is None or node['key'] == key:
        return node
    
    if key < node['key']:
        return search_node(node['left'], key)
    else:
        return search_node(node['right'], key)

def get_successor(node):
    if node is None:
        return None
    
    if node['right'] is not None:
        return get_minimum(node['right'])
    
    y = node['p']
    while y is not None and node == y['right']:
        node = y
        y = y['p']
        
    return y

def get_minimum(node):
    while node['left'] is not None:
        node = node['left']
        
    return node
```

```python
def to_tsv(node, data, file):
    if node is None:
        return
    
    to_tsv(node['left'], data, file)
    
    # 현재 노드의 키 값으로 데이터에서 검색하여 해당되는 값을 변수에 저장
    node_data = data.get(node['key'])
    if node_data:
        # 해당하는 값들을 tsv 파일로 저장
        file.write("{}\t{}\t{}\t{}\t{}\t{}\n".format(node['key'], node_data['company_name'], node_data['address'], node_data['zip'], node_data['phone'], node_data['email']))
    
    to_tsv(node['right'], data, file)


with open('address_book2020_copy.tsv', 'w') as file:
    file.write("name\tcompany_name\taddress\tzip\tphone\temail\n")
    to_tsv(tree['root'], data, file)
```

이진검색트리를 판별하는 함수- 중간고사 대비

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def is_bst_helper(node, min_val, max_val):
    if node is None:
        return True
    
    if node.val <= min_val or node.val >= max_val:
        return False
    
    return is_bst_helper(node.left, min_val, node.val) and is_bst_helper(node.right, node.val, max_val)

def is_bst(root):
    return is_bst_helper(root, float('-inf'), float('inf'))

# 이진트리 생성 예시
root = TreeNode(5)
root.left = TreeNode(9)
root.right = TreeNode(7)
root.left.left = TreeNode(1)
root.left.right = TreeNode(4)
root.right.left = TreeNode(6)
root.right.right = TreeNode(8)

# 이진검색트리 여부 판별
is_bst(root)  # True
```
